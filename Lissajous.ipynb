```python
import numpy as np
import matplotlib.pyplot as plt


t = np.linspace(0,15.100)

w1=1
w2=np.array([1,2,3])

k=0
fig = plt.figure(figsize=(10,6))
for i in range(3):
  for j in range(5):
    x = np.sin(w1*t+np.pi/4*j)
    y = np.cos(w2[i]*t)
    plt.subplot(3,5,k+1)
    plt.plot(x,y)
    k+=1

plt.show()
```
